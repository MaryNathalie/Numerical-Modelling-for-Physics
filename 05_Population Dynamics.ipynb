```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logistics = lambda r,x: 4*r*x*(1-x)
#x_n is proportional to the population density of a species at time n.
#Its growth is described by 4rx_n (by itself, exponential growth i.e. population explosion for r>1/4)
#Growth is lmited by the nonlinear contribution (-4rx^2) limited food supply
x = 0.65
r = 0.87
i = range(41)
xlist = []

for ii in i:
    xlist.append(x)
    x = logistics(r,x)

plt.title('Iteration starting from 'r'$x_0 = 0.65$'' using the parameter r = 0.87. \n The 'r'$x_n$'
          ' values move towards an attractor of period 4.')
plt.xticks(range(0,41,10))
plt.xlabel('n')
plt.ylabel('x(n)')
plt.plot(i,xlist,'--bo')
plt.show()
print('The x_n values move towaard an attractor of period 4, which is described by the fixed points of'
      '\nthe fourfold iterated functions.')
```

```python
x = np.linspace(0,1,1000)
xx = x
r = 0.87

for i in range(4):
    if i !=2 :
        plt.plot(xx,logistics(r,x),label=str(i+1))
    x = logistics(r,x)
plt.title('The function 'r'$f(x)$'' and its two- and fourfold iterations. \n' 
          'The latter has eight fixed points.')
plt.plot([0,1],[0,1],label='x = y')
plt.legend()
plt.xlabel('x')
plt.show()
print('The fourfold iterated function intersects the straight line y=x 8 times, but only four of those are stable fixed points.')
```

```python
i = 1000
r = np.linspace(0,1,i)
x = np.linspace(0,1,i)

for ii in range(i-1): #0-999
    x = logistics(r,x)
    if ii >= 100: #not showing 0 to 99
        plt.scatter(r,x,s=0.0001)
        
plt.title('Each vertical line contains 1000 'r'$x_n$'' values from the iterated \n parabola, '
          'not showing the first 100 points. The parameter 'r'$r$'' \n varies between 0 and 1 along the horizontal.')
plt.xticks([0,0.5,1.0])
plt.xlabel('r')
plt.ylabel('x')
plt.show()  
print('As r is increased, attractors with period 8,16,32,... are generated until the transition to chaos occurs at r = 0.89....'
     '\nThis is visible in the figures, in which 1000 x_n values are plotted for each value of r, after an initial transition'
      '\nperiod of about 100 steps. Thus, attractors with period p are visible as p points, whereas chaotic orbits fill several')

for ii in range(i-1): #0-999
    x = logistics(r,x)
    if ii >= 100: #not showing 0 to 99
        plt.xlim(0.88,1)
        plt.scatter(r,x,s=0.004)
plt.title('A close-up for 'r'$r$'' between 0.88 and 1')
plt.yticks([0,0.5,1.0])
plt.xticks([0.88,0.94,1.0])
plt.xlabel('r')
plt.ylabel('x')
plt.show()

print('In the chaotic region , one sees many windows with periodic orbits. The largest window has a period of 3. In this'
     '\nwindow, there is a period doubling again: as r increases, one obtains the periods 3,6,12,24.... until a chaotic region'
     '\nappears. Indeed, it can be shown that all period occurs. ')
```

```python
i = 1000
r = 0.9342
x = np.linspace(0.65,1,i)
xlist = []

for ii in range(1000): #0-999
    x = logistics(r,x)
    xlist.append(x)

plt.title('Frequency distribution of the 'r'$x_n$'' values in the iteration\n of the parabola for a parameter'
          'value 'r'$r = 0.934$'', just below the\n window with period 5.')
plt.hist(x,250)
plt.show()
print('This shows the density of x_n values. Periodic orbits with period p show up as a series of lines with p peaks, whereas'
     '\nchaotic orbits jump back and forth between various bands, each with continuous densities. This figure shows the chaotic'
     '\nmotion below a five-cycle; though there is just one band there, one still sees the 5 peaks of the adjacent cycle.')
```

```python
def plot(r,x,n,c):
    xx = np.linspace(0,1,100)
    plt.plot(xx, logistics(r,xx))
    plt.plot([0,1],[0,1])
    xlist = []     #(x     f(x)  f(x)   ff(x),...)
    ylist = []     #(f(x)  f(x)  ff(x)  ff(x),...)
    for i in range(n):     #for every i, 2 point for x list and 2 points for ylist are added
        xlist.append(x)     
        x = logistics(r,x)
        ylist.append(x)
        ylist.append(x)
        xlist.append(x)
    plt.plot(xlist,ylist,linewidth=c)

plot(0.87460,0.5,5,0.5) #5 (x, f(x)) and (f(x),f(x))
plt.title('Superstable four-cycle with 'r'$r=0.874640$')
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel(r'$x_n$')
plt.ylabel(r'$x_{n+1}$')
plt.show()
print('Superstable periodic orbits start at x_0 = 1/2 and return after 2^l steps because fprime(1/2)=0 so deviations'
     '\nconverge towards the 2^l cycle very quickly.')

plot(0.9642,0.5,1000,0.1)
plt.title('Three chaotic bands for 'r'$r=0.9642$'', \njust above the large window with the three-cycle')
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel(r'$x_n$')
plt.ylabel(r'$x_{n+1}$')
plt.show()
```
