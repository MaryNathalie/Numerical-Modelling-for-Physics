```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
qn = lambda q,p: (q+p)/2
n = 50000
# set random numbers [0,40)
a = 0
b = 500
p = (b-a)*np.random.randn(3,2)+a
q = np.array([0,0])
xlist = []
ylist = []
for i in range(n):
    r = np.random.randint(3)
    q = qn(q,p[r])
    xlist.append(q[0])
    ylist.append(q[1])
colors = cm.rainbow(np.linspace(0, 1, len(xlist)))
plt.scatter(xlist,ylist,s=0.01,color=colors)

#print(q)
print('Random motion of leads to an obviously regular structure consisting of triangle nested inside each other.\n'
     'All triangles look the same, independent of their size.\n'
     'Structure is nearly empty, as each triangle has an inf number of holes at any scale.\n')
```

```python
D = spatial dimension
L = linear dimension
M = mass has N number of components
N = number of particles

D Fractal = log 3/log 2 = 1.58
Sierpinski gasket
**M proportional to L^D
M2/M1 = (2L/L)^D = 2^D
D cube = 3
D square = 2 
D line = 1
**M proportional to N
$D = lim(L->inf) log M/log L

Mt = 3^(-t)M
Lt = 2^(-t)L
D = lim(t->inf) = ln Mt/ln Lt

Cover object with small cubes with edges eps
%N(eps) = smallest no. of cubes needed
**N(eps) proportional to eps^(-D)
$D = -lim(eps->0) log N(eps)/log (eps)
To cover
N box = (L/eps)^3
N square = (L/eps)^2
N line = L/eps

FRACTAL DOES NOT AGREE
Border of Map
%eps = ruler of length 
%N(eps) = steps
$L = eps N(eps)
D border = 1
smaller eps: more steps than L/eps: D not= 1
** L depends on eps
A border with a fractal with 1<D<2 
** L prorportional to eps^(1-D) where L diverges as eps -> 0
Borders do not have defined length

Polymer molecule
Assume: distance between adjacent monomers = constant; angles between 3 consec. monomers = independent, randomly dist.
Embedding dimension d  = dimension where obj is located = random motion in d-dimensional space
```

```python
p = np.array(([0,0], [1,0],[0.5,0.5]))

def tri(p, count = 0, iterate = 5):
    if count < iterate:
        count += 1
        for i in range(len(p)):
            corner = [p[i]]
            for j in range(len(p)):
                x1 = qn(p[i], p[j])
                corner.append(x1)
                z = list(zip(p[i], x1))
                plt.scatter(z[0], z[1])
            tri(corner, count)

tri(p)
```
